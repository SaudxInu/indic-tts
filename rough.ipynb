```python
# pip install numpy vllm
import time

import numpy as np
from vllm import LLM, SamplingParams


def create_llm():
    return LLM(
        model="snorbyte/snorTTS-Indic-v0",
        quantization="fp8",
    )


def get_sampling_params():
    return SamplingParams(
        temperature=0.4,
        top_p=0.9,
        max_tokens=2048,
        repetition_penalty=1.05,
        stop_token_ids=[128258],
    )
```

```python
llm = create_llm()
sampling_params = get_sampling_params()
```

```python
def run(prompt, llm, sampling_params):
    num_words = len(prompt[0].split(" "))
    start = time.perf_counter()
    outputs = llm.generate(prompt, sampling_params)
    end = time.perf_counter()
    elapsed_time = end - start
    num_input_tokens = len([token_id for token_id in outputs[0].prompt_token_ids])
    num_output_tokens = len([token_id for token_id in outputs[0].outputs[0].token_ids])
    return {
        "num_words": num_words,
        "num_input_tokens": num_input_tokens,
        "num_output_tokens": num_output_tokens,
        "time": elapsed_time,
    }
```

```python
for _ in range(10):
    row = data[0]

    prompt = [
        f"<custom_token_3><|begin_of_text|>{row['language']}{row['user_id']}: {row['utterance']}<|eot_id|><custom_token_4><custom_token_5><custom_token_1>"
    ]

    _ = run(prompt, llm, sampling_params)
```

```python
logs = []
for _ in range(5):
    for row in data:
        prompt = [
            f"<custom_token_3><|begin_of_text|>{row['language']}{row['user_id']}: {row['utterance']}<|eot_id|><custom_token_4><custom_token_5><custom_token_1>"
        ]

        log = run(prompt, llm, sampling_params)

        logs.append(log)
```
